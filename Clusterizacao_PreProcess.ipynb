### Separação de textos

```python
dados.Interests[1].replace(" and ", " ")
```

```python
dados.Interests = [i.replace(" and " , "_") for i in dados.Interests]
```

```python
list_separed = list()
list_iters = list()
for i in dados.Interests:
    list_separed.append([j.strip()[1:-1].upper() for j in i.split(",")])
    for j in i.split(","):
        list_iters.append(j.strip()[1:-1].upper().replace(" ", "_"))
    
```

```python
list_iters = set(list_iters)
```

```python
list_iters
```

```python
set_interesses = [set(i) for i in list_separed]
```

```python
list_separed[0][0]
```

```python
interesses = pd.DataFrame(columns= list(list_iters))
```

### Binarização

```python
list_dicts = list()
for conj_interesse in list_separed:
    dict_interesse = dict()
    for interesse in conj_interesse:
        for coluna in interesses.columns.values:
            if interesse == coluna:
                dict_interesse[coluna] = 1
            else:
                dict_interesse[coluna] = 0
    list_dicts.append(dict_interesse)
```

```python
interesses = pd.DataFrame(list_dicts)
```

### Esquema do dataset final

```python
dados_final = dados.reset_index().merge(interesses.reset_index())
```

```python
dados_final.drop(["index", "Interests"], axis = 1, inplace=True)
```

```python
dados_final.drop_duplicates("Name", inplace=True)
```

```python
dados_final
```

```python
columns = dados_final.columns.values
```

### Exportação do JSON

```python
import tqdm
```

```python
dict_results = dict()
for col in tqdm.tqdm(columns[2:]):
    nomes = dados_final.iloc[:20000].query(f"{col} == 1").Name.values
    list_relacionamento = list()
    for tup in product(nomes, repeat=2):
        if tup[0] != tup[1]:
            list_relacionamento.append([tup[0], tup[1]])
    dict_results[col] = list_relacionamento
```

```python
from __future__ import print_function
import json

open('dicionario_final.json','w').write(json.dumps(dict_results))
```
